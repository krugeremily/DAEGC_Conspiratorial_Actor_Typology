# Imports and Data Loading

```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
full_data = pd.read_csv('../results/author_full_features_and_clusters.csv.gzip', compression='gzip')
```

```python
full_data.columns
```
